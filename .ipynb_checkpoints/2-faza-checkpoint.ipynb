```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn import linear_model as lm
from datetime import datetime
import json
from sklearn import preprocessing
from sklearn.preprocessing import Imputer
from scipy.stats import boxcox
import seaborn as sns
import category_encoders as ce
from sklearn.neighbors import KNeighborsClassifier
```

## Odstránenie nekonzistencie dát

V metóde process sme pridali úpravy atribútov z predchádzajúcej časti. Všetky zmeny vykonávané v datasetoch vieme zavolať jednou metódou. 

```python
def process(data):
    data.loc[(data.age < 0 ),'age'] *= -1
    data.loc[(data.age > 1000 ),'age'] /= 100
    
    data['sex'] = data['sex'].replace(['f'], 'F')
    data['sex'] = data['sex'].replace(['m'], 'M')
    data['sex'] = data['sex'].replace(['?'], np.nan)
    data['query on thyroxine'] = data['query on thyroxine'].replace(['F','FALSE'], 'f')
    data['query on thyroxine'] = data['query on thyroxine'].replace(['T'], 't')
    data['FTI'] = data['FTI'].replace(['?'], np.NaN)
    data['FTI']=data['FTI'].astype(float)
    data = data.drop('TBG measured', axis=1)
    data=data.drop(['TBG'],axis=1)
    data['class']=data['class'].str.split('.', expand=True)[0]
    
    data['education']=data['education'].replace([' HS_grad'], ' HS-grad')
    data['education']=data['education'].replace([' Some_college'], ' Some-college')
    data['education']=data['education'].replace([' Assoc_voc'], ' Assoc-voc')
    data['education']=data['education'].replace([' Assoc_acdm'], ' Assoc-acdm')
    data['education']=data['education'].replace(['Assoc_acdm'], ' Assoc_acdm')
    data['education']=data['education'].replace([' 7th_8th'], ' 7th-8th')
    data['education']=data['education'].replace([' Prof_school'], ' Prof-school')
    data['education']=data['education'].replace([' 5th_6th'], ' 5th-6th')
    data['native-country'] = data['native-country'].replace([' ?'], np.nan)
    
    data['workclass'] = data['workclass'].replace([' ?','??'], np.nan)
    data['occupation'] = data['occupation'].replace([' ?'], np.nan)
    data['date_of_birth'].dropna(axis=0, how='all', inplace= True)
    index_stlpca_date = 30

    for i in range(len(data['date_of_birth'])):
        data.iat[i,index_stlpca_date] = data.iat[i,index_stlpca_date].replace("/", "-")
        date_split = data.iat[i,index_stlpca_date].split('-')

        if(len(date_split[0]) == 2 & len(date_split[2]) == 2):    
            novy_datum = str()
            novy_datum = '19' + date_split[0]
            data.iat[i,index_stlpca_date] = novy_datum + "-" + date_split[1] + "-" + date_split[2]

        if(len(date_split[2]) == 4):
            data.iat[i,index_stlpca_date] = date_split[2] + "-" + date_split[1] + "-" + date_split[0]

        if(len(date_split[2]) > 4):
          if(date_split[2].count(':') != 2):
                rozdelene_medzerou = date_split[2].split(' ')
                date_split[2] = rozdelene_medzerou[0] + ' ' + rozdelene_medzerou[1] + ':' + rozdelene_medzerou[2] + ':' + rozdelene_medzerou[3] 
                data.iat[i,index_stlpca_date] = date_split[0] + "-" + date_split[1] + "-" + date_split[2]

        try:
          d = datetime.strptime(data.iat[i,index_stlpca_date], "%Y-%m-%d")
          data.iat[i,index_stlpca_date] = d

        except ValueError:
          continue
    
    data['date_of_birth'] = data['date_of_birth'].astype("datetime64[ns]")
    
    medical_info = pd.DataFrame(list(map(lambda x: json.loads(x.replace("'", "\"")), data['medical_info'])))
    data=pd.concat([data,medical_info], axis=1, join_axes=[data.index])
    data= data.drop(['medical_info'],1)
    
    return data
```

## Vytvorenie percentilov a nahradenie outlierov

```python
def vypocet_percentilov(array,dolper,horper):
  dole=np.percentile(array, dolper)
  hore=np.percentile(array, horper)

  print(dole)
  print(hore)

  array= np.where(array<=dole,dole,array)
  array= np.where(array>=hore,hore,array)
   

  return array,dole,hore
  
```

V tejto metóde nahradzujeme vychýlené hodnoty smerom nahor a nadol. Vstupné parametre dolper a horper určujú hranice percentilov. Hodnoty menšie ako je dolper percentil nahradzujeme tou hodnotou  percentila. Analogicky to funguje aj pri hornej hranici. Metóda vracia upravené hodnoty v poli. Vo výpise nám vypíše dolný a horný percentil. 

## Lineárna regresia

Metóda na nahradzovanie chýbajúcich hodnôt pomocou lineárnej regresie. Funkcia vracia natrenovaný model, aby sme rovnaký model použili na testovacom datasete. Rovnako aj ostatné funkcie vracajú svoje modeli(knn,medián a pod)

```python
def lin_regresia(array,column):
  regr = lm.LinearRegression()
  nan_mask=np.isnan(array)
  not_nan_mask = ~np.isnan(array)
  arrayage=np.array(column)

  arrayagetrain= arrayage[not_nan_mask]
  arraytrain= array[not_nan_mask]

  arrayagetrain= arrayagetrain.reshape(1,-1).transpose()
  arraytrain= arraytrain.reshape(1,-1).transpose()


  arrayagepredict= arrayage[nan_mask]
  arrayagepredict=arrayagepredict.reshape(1,-1).transpose()

  regr.fit(arrayagetrain, arraytrain )

  a = regr.predict(arrayagepredict)
  
  j=0
  for i in range(len(array)):
    if(np.isnan(array[i])): 
      array[i]=a[j]
      j=j+1
  
  return array,regr
  
```

## KNN

```python
def knn(predict,train):

  array=np.array(predict)

  aa=predict.isnull()
  nan_mask=np.array(aa)
  not_nan_mask = ~nan_mask

  arrayage=np.array(train)

  arrayagetrain= arrayage[not_nan_mask]
  arraytrain= array[not_nan_mask]

  arrayagetrain= arrayagetrain.reshape(1,-1).transpose()
  arraytrain= arraytrain.reshape(1,-1).transpose()


  arrayagepredict= arrayage[nan_mask]
  arrayagepredict=arrayagepredict.reshape(1,-1).transpose()

  neigh = KNeighborsClassifier(n_neighbors=5)

  neigh.fit(arrayagetrain, arraytrain) 

  a=neigh.predict(arrayagepredict)

  j=0
  for i in range(len(nan_mask)):
    if(nan_mask[i]): 
      array[i]=a[j]
      j=j+1
  
 
  return array,neigh
```

Metóda nahradzovania NaN hodnôt pomocou algoritmu KNN. Túto techniku používame najmä pre kategorické atribúty.

## Transformácia

Transformácia atribútu, ktorý sa posiela ako parameter do metódy. Na transformáciu používame boxcox. Najprv však musíme vybrať len nie NaN hodnoty, aby sme mohli vykonať trasformáciu. Metóda vráti upravené polia s hodnotami NaN a bez. Metóda vypisuje tiež počet NaN hodnôt v atribúte.

```python
def transformacia(attribute):
  array=np.array(data[attribute])
  not_nan_mask = ~np.isnan(array)
  transformed, att = boxcox(array[not_nan_mask])
  #transformed = normalization(transformed, np.mean(transformed), np.std(transformed))
  #transformed =preprocessing.scale(transformed)

  array[not_nan_mask]= transformed

  transf=pd.Series(array)
  print(transf.isnull().sum())

  _ = plt.figure(figsize=(10,5))
  _ = sns.kdeplot(transf, shade=True)
  
  # #TRANSFORMED BEZ NAN !!!!
  # #ARRAY Z NAN !!!!!

  
  return transformed, array, att
```

Musíme vrátiť atribút, aby sme mohli rovnakú transformáciu vykonať aj na testovacom datasete.

## Výmena NaN mediánom

Metóda na výmenu chýbajúcich hodnôt mediánom. Vracia nám pole s nahradenými NaN hodnotami.

```python
def vymenananmedian(transformed,array):
  median =np.median(transformed)
  print(median)
  array= np.where(np.isnan(array) ,median,array)
  transf=pd.Series(array)
  plt.figure(figsize=(10,5))
  sns.kdeplot(transf, shade=True)
  
  return array,median
```

## Výmena NaN mediánom segmentu

Metóda na výmenu chýbajúcich hodnôt mediánom segmentu. Vracia nám pole s upravenými hodnotami. Ako segmenty využívame atribút sex

```python
def vymenananmediansegmentu(transformed,array):
  
  median =np.median(transformed)
  df= pd.DataFrame(array)

  df2=pd.concat([df,data['sex']],axis=1, join_axes=[df.index])
  df2.columns = ['value', 'sex']
  
  medianseg = df2.groupby('sex').aggregate('median')
  medianseg['value']['nezname']=median

  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('F'),'value'] = medianseg.at['F','value']
  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('M'),'value'] = medianseg.at['M','value']
  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('nezname'),'value'] = medianseg.at['nezname','value']
  
  print(df2['value'].isnull().sum())
  array= np.array(df2['value'])
    
  return array,medianseg
```

# Asociačné pravidlá

## Načítanie a úprava testovacieho datasetu

V tejto časti sú funkcie, ktoré vymenia hodnoty v datasete na základe naučených hodnôt z trénovacieho datasetu.

```python
def transformacia2(attribute,att):
  array=np.array(datatest[attribute])
  not_nan_mask = ~np.isnan(array)
  transformed= boxcox(array[not_nan_mask],lmbda=att,alpha=None)
  #transformed = normalization(transformed, np.mean(transformed), np.std(transformed))
  #transformed =preprocessing.scale(transformed)

  array[not_nan_mask]= transformed

 
  # #TRANSFORMED BEZ NAN !!!!
  # #ARRAY Z NAN !!!!!

  
  return transformed, array
```

Do parametru funkcie vkladáme aj názov atribútu, aby vykonala rovnaká transformácia ako v trénovacom datasete.

```python
def vymenananmedian2(array,median):
  array= np.where(np.isnan(array) ,median,array)
  transf=pd.Series(array)
  
  return array
```

```python
def vypocet_percentilov2(array,dole,hore):
  
  array= np.where(array<=dole,dole,array)
  array= np.where(array>=hore,hore,array)
   

  return array
```

```python
def lin_regresia2(array,column,regr):
  
  nan_mask=np.isnan(array)
  not_nan_mask = ~np.isnan(array)
  arrayage=np.array(column)

  arrayagepredict= arrayage[nan_mask]
  arrayagepredict=arrayagepredict.reshape(1,-1).transpose()

  a = regr.predict(arrayagepredict)
  
  j=0
  for i in range(len(array)):
    if(np.isnan(array[i])): 
      array[i]=a[j]
      j=j+1
  
  return array
```

```python
def vymenananmediansegmentu2(array,medianseg):
 
  df= pd.DataFrame(array)

  df2=pd.concat([df,datatest['sex']],axis=1, join_axes=[df.index])
  df2.columns = ['value', 'sex']
  
  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('F'),'value'] = medianseg.at['F','value']
  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('M'),'value'] = medianseg.at['M','value']
  df2.loc[df2['value'].isnull() & df2['sex'].str.contains('nezname'),'value'] = medianseg.at['nezname','value']
  
  array= np.array(df2['value'])
    
  return array
```

## Transformacia testovacieho datasetu,rovnake pravidla ako v trenovacom datasete

```python
def transformuj(data):
  
  data['sex'] = data['sex'].replace(np.nan, 'nezname')
  
  transformed_age, array_age = transformacia2('age',attage)
  array_age = vymenananmedian2(array_age,median_age)
  array_age = vypocet_percentilov2(array_age,minage,maxage)
  data['age']=pd.DataFrame(array_age)
  
  transformed_TSH, array_TSH = transformacia2('TSH',attTSH)
  array_TSH=vymenananmediansegmentu2(array_TSH,median_TSH)
  array_TSH = vypocet_percentilov2(array_TSH,minTSH,maxTSH)
  data['TSH']=pd.DataFrame(array_TSH)
  
  transformedT3,arrayT3 = transformacia2('T3',attT3)
  arrayT3=lin_regresia2(arrayT3,data['age'],lr_T3)
  arrayT3 = vypocet_percentilov2(arrayT3,minT3,maxT3)
  data['T3']=pd.DataFrame(arrayT3)
  
  transformed_TT4,array_TT4 = transformacia2('TT4',attTT4)
  array_TT4=lin_regresia2(array_TT4,data['T3'],lr_TT4)
  array_TT4 = vypocet_percentilov2(array_TT4,minTT4,maxTT4)
  data['TT4']=pd.DataFrame(array_TT4)
  
  transformed_T4U,array_T4U = transformacia2('T4U',attT4U)
  array_T4U=lin_regresia2(array_T4U,data['T3'],lr_T4U)
  array_T4U = vypocet_percentilov2(array_T4U,minT4U,maxT4U)
  data['T4U']=pd.DataFrame(array_T4U)
  
  transformed_FTI,array_FTI = transformacia2('FTI',attFTI)
  array_FTI=lin_regresia2(array_FTI,data['TT4'],lr_FTI)
  array_FTI = vypocet_percentilov2(array_FTI,minFTI,maxFTI)
  data['FTI']=pd.DataFrame(array_FTI)
  
  transformed_fnlwgt, array_fnlwgt = transformacia2('fnlwgt',attfnlwgt)
  array_fnlwgt=vymenananmediansegmentu2(array_fnlwgt,median_fnlwgt)
  array_fnlwgt = vypocet_percentilov2(array_fnlwgt,minfnlwgt,maxfnlwgt)
  data['fnlwgt']=pd.DataFrame(array_fnlwgt)
  
  
  #daval som iba tie ktore sa pouzivaju v pravidlach, nie su vsetky atributy
  
  return data
```
